# The Impact of Technology on Mental Health: Insights from 2024

In today's digital age, technology is an ever-present force in our daily lives. From smartphones and social media to office tools and entertainment, our dependence on technology is greater than ever before. While technology brings many benefits, it also raises important questions about its impact on our general well-being, particularly our mental health. 

In this analysis, we explore the relationship between technology usage and mental health, drawing insights from the 2024 Mental Health and Technology Usage dataset from Kaggle. By examining patterns and correlations in the data, we aim to uncover how different forms of technology influence stress, anxiety, and overall mental health.

### Exploring Key Questions / Visualizations

```python
#set consistent plot style
sns.set(style="whitegrid")

#how does the amount of technology usage correlate with reported stress levels?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Technology_Usage_Hours', y='Stress_Level', alpha=0.7)
plt.title('Correlation between Technology Usage and Stress Level')
plt.xlabel('Technology Usage (hours/day)')
plt.ylabel('Stress Level')
plt.show()
```

This graph shows the relationship between daily technology usage (in hours) and reported stress levels, which are categorized as Low, Medium, and High. Interestingly, the data points are evenly spread across all stress levels, regardless of how many hours people spend using technology. This suggests that simply spending more time on technology does not necessarily lead to higher stress levels. Instead, individuals with similar technology usage can experience different levels of stress, indicating that other factors, such as how technology is used or personal circumstances, might play a significant role in determining stress levels.

```python
#are younger people more affected by technology in terms of their mental health compared to older people?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age', y='Stress_Level', palette='coolwarm')
plt.title('Stress Levels across Different Ages')
plt.xlabel('Age')
plt.ylabel('Stress Level')
plt.show()
```

This box plot illustrates the distribution of stress levels across different age groups, categorized into Low, Medium, and High stress. The plot shows that individuals reporting low, medium, or high stress levels span a broad age range, from their early 20s to late 50s, with no single age group dominating any specific stress category. This suggests that age alone may not be a strong determinant of stress levels, as people across different age groups can experience similar levels of stress. The even distribution across ages suggests the need to consider other factors, such as lifestyle or work environment, when addressing stress, rather than attributing it solely to age.

```python
#what types of technology usage are most associated with mental health issues?
plt.figure(figsize=(12, 8))
usage_types = ['Social_Media_Usage_Hours', 'Gaming_Hours', 'Screen_Time_Hours']
average_stress = df.groupby('Stress_Level')[usage_types].mean().T

average_stress.plot(kind='bar', figsize=(12, 8), colormap='coolwarm')
plt.title('Average Technology Usage Hours by Stress Level')
plt.xlabel('Type of Technology Usage')
plt.ylabel('Average Hours per Day')
plt.xticks(rotation=45)
plt.legend(title='Stress Level', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

This bar chart illustrates the average hours spent on social media, gaming, and overall screen time, categorized by stress levels—High, Medium, and Low. The data shows that screen time is consistent across all stress levels, indicating it is a significant part of daily life regardless of how stressed individuals feel. Social media usage is also stable across different stress levels, suggesting that it is a common activity regardless of stress. Interestingly, gaming hours are nearly identical across all stress levels, indicating that gaming does not appear to be strongly associated with differences in stress levels. Overall, the chart suggests that none of these types of technology usage show a significant variation based on stress levels.

```python
#heatmap of numerical variables
plt.figure(figsize=(12, 8))
correlation_matrix = df[['Age', 'Technology_Usage_Hours', 'Social_Media_Usage_Hours', 'Gaming_Hours', 'Screen_Time_Hours', 'Sleep_Hours', 'Physical_Activity_Hours']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

This correlation heatmap visualizes the relationships between different variables in the dataset, such as age, technology usage, social media usage, gaming hours, screen time, sleep hours, and physical activity hours. The colors range from deep blue, indicating little to no correlation, to red, which indicates a stronger positive correlation. The map shows that there are generally weak correlations between the variables, with most squares displaying a blue color. This suggests that these variables are independent of one another, meaning, for example, that an increase in technology usage does not necessarily predict a decrease in physical activity or sleep hours.

```python
df['Stress_Level'] = df['Stress_Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
df['Mental_Health_Status'] = df['Mental_Health_Status'].map({'Excellent':0, 'Fair':1, 'Good': 2, 'Poor': 3})
df['Support_Systems_Access'] = df['Support_Systems_Access'].map({'No': 0, 'Yes': 1})
df['Work_Environment_Impact'] = df['Work_Environment_Impact'].map({'Negative': -1, 'Neutral': 0, 'Positive': 1})
df['Online_Support_Usage'] = df['Online_Support_Usage'].map({'Yes': 1, 'No': 0})

# Then rerun the correlation analysis:
numeric_df = df.select_dtypes(include=['float64', 'int64'])
stress_corr = numeric_df.corr()['Stress_Level'].sort_values(ascending=False)
mental_health_corr = numeric_df.corr()['Mental_Health_Status'].sort_values(ascending=False)

print("Correlation with Stress Level:\n", stress_corr)
print("\nCorrelation with Mental Health Status:\n", mental_health_corr)
```

The revised correlation analysis shows that the relationships between various factors and both stress levels and mental health status remain weak, with all correlation coefficients still close to zero. The factor most correlated with stress level is Physical Activity Hours (0.0157), though this is still a very weak relationship, indicating that physical activity has a minimal impact on stress. Other factors like Social Media Usage Hours, Support Systems Access, and Sleep Hours show slightly positive but very weak correlations with stress levels, suggesting they contribute little to changes in stress levels.

For mental health status, Support Systems Access shows the highest correlation (0.0239), which is still weak, indicating a minor positive relationship where better support systems might be associated with slightly better mental health. Similarly, Technology Usage Hours has a very weak positive correlation (0.0203) with mental health status. Negative correlations are observed with Age and Physical Activity Hours, but these are also minimal.

### Multifactorial Influence Hypothesis

Stress levels and mental health status are influenced by a combination of multiple factors rather than any single factor alone. While no individual factor shows a strong correlation, the cumulative effect of several weakly correlated factors could be significant.
